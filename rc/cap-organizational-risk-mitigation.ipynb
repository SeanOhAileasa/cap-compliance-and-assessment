###### Risk Identification Process
***

Every enterprise needs a method/team responsible for calculation and determining their risk appetite. 

> Enterprise Risk Management (``ERM``) - Calculation / Determining Risk Appetite

This is known as the Enterprise Risk Management (``ERM``) - some examples of why to establish an ``ERM`` are as follows:

- Keep confidential information confidential.


- Keep trade secrets from the public eye.


- Avoid damaged reputation and financial losses.


- Avoid legal issues.


- Ensure business operations remain strong in the long term.


- Establish trust in your business relationships.


- Make your shareholders and stakeholders happy.

> ``NIST SP 800-39`` - Categories for Identifying Risk

Lays out a great starting point for establishing risk in its ``Special Publication 800-39``, where they depict four major categories for identifying risk:

- Frame

Take a top-down approach to ensure decision makers are able to make decisions on all areas of risk.

- Assess

Identify and prioritize workflows within the business.

- Respond

Effectively fix and remediate any areas of concern.

- Monitor

Consistently evaluate the effectiveness of our risk plan to determine if it needs to be altered or changed.

> Risk Management Framework - Terms

Terms in mind when applying the Risk Management Framework:

- Legal Perspective

There are legal obligations to adhere to, where the fine of not complying would be significant.

- Reputational

The toll it would take on your further sales would drastically reduce the overall value of the company.

- Business Continuity

A system that would take down or halt business operations if it were to be compromised.

###### Systems Assessments
***

> Security Doesnt Make Money

Rather, it is an insurance plan that protects an enterprise's assets from future concerns. 

In order to prioritize assets that require more security, their value must be assessed against the business. 

System assessments are completed by compiling all the tangible and intangible assets that support them - these include:

- People


- Buildings, equipment and paper documents


- Ideas, reputation and brand name


- Procedures (supply chains or operating steps)

> Mission Essential Functions (``MEF``)

Based on this list created, should be able to establish your Mission Essential Functions (``MEF``). 

``MEF`` is a business or organizational activity that is too critical to be deferred for anything more than a few hours.

- Other factors to consider in determining the ``MEF`` are threats and vulnerabilities that may affect the assets involved.

###### Business Impact Analysis - ``BIA``
***

Business Impact Analysis (``BIA``) is a systematic activity that identifies organizational risks and determines their effect on ongoing mission critical operations.

``BIA`` is calculated by using different metrics to determine system availability as follows:

> **Metrics**:

- Maximum Tolerable Downtime - ``MTD``

The longest period of time a business can be inoperable without causing irrevocable business failure or damage.

- Recovery Time Objective - ``RTO``

The length of time it takes after an event to resume normal business operations.

- Work Recovery Time - ``WRT``

The additional time to the ``RTO`` of individual systems to perform reintegration and testing of a restored or upgraded system following an event.

- Recovery Point Objective - ``RPO``

The longest period of time that an organization can tolerate lost data being unrecoverable.

###### Risk Prioritization
***

Regarding Risk Prioritization, there are several important definitions:

- Risk Mitigation

The response of reducing risk to fit within an organization's risk appetite.

- Risk Deterrence

The response of deploying security controls to reduce the likelihood and impact of a threat.

- Risk Avoidance

The practice of ceasing activity that presents risk.

- Risk Transference

The response of moving or sharing the responsibility of risk to another entity.

- Risk Acceptance

The response of determining that a risk is within the organization's appetite and no countermeasures other than ongoing monitoring is required.

###### Security Control Prioritization
***

Determining Security Controls and when to implement them will be based on several factors, like whether it is required by a framework or regulatory body, the level of risk required, or even the cost of implementing a control.

- Return on Security Investment - ``ROSI``

A metric to calculate whether a security control is worth the cost of deploying and maintaining it.

- Annual Loss Expectancy - ``ALE``

The total cost of a risk to an organization on an annual basis - this is determined by multiplying the Single Loss Expectancy (``SLE``) by the Annual Rate of Occurrence (``ARO``).

$$\text{(ALE - ALE}^{M}\text{)}-\text{C / C = ROSI}$$

``ROSI`` can also be calculated as follows:

$$\text{(ALE}\times\text{MR}\text{)}-\text{C / C = ROSI}$$

- Engineering Tradeoff

In Risk Prioritization, an ``Engineering Tradeoff`` is an assessment of the benefit of risk reduction against the increased complexity or cost in a system design or specification.

###### Communicating Risk
***

There are several ways that risk can be communicated within an organization - some of these ways may be required, whereas others are highly recommended.

Here are some standard ways that risks are communicated:

> Risk Register

A document that highlights the results of a risk assessment in an easily comprehensible manner. 

Its purpose is for managers and technicians to understand risks associated with the workflows that they manage.

> Compensating Controls

A security measure that takes on risk mitigation when a primary control fails or cannot completely meet the expectations.

Also important to document any exceptions that are given within the environment.

- These should track the business purpose, reason for the exception, duration it will last for, and all personnel involved with the exception.

###### Training for Risk
***

Risk is not just spreadsheets and calculations. 

Like all difficult things, it must be practiced in order to ensure its completeness. 

The following are a few common ways risk management can be practiced:

> Tabletop Exercises

A discussion of simulated emergency situations and security incidents that could arise.

> Penetration Test

A test that uses active tools and utilities to evaluate security by simulating an attack on a system or network. 

- A pentest will:

1. Verify that a threat exists.


2. Actively test and bypass security controls.


3. Exploit vulnerabilities on the system.
